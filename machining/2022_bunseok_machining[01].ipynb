#### 학습목표
AI란?
- 기계학습(machin-learning) : framework(sklearn , tensorflow , caffee, pytorch, etc...)
- ANN 딥러닝(deep-learning) : ANN(CNN - 이미지에 대한 딥한 학습 , RNN - 자연어에 대한 딥한 학습) 
- AI란 학습(데이터)을 통한 결과를 리턴하는 로봇이다.
- 분석(통계 , 시각화) - 서비스화를 위해서는 로봇 <- 데이터 주입해서 학습된 모델이 필요하다.
- 데이터가 쓰레기면 로봇도 쓰레기가 된다. (garbage in -> garbage out)
- 인공지능에서 가장 중요한건 데이터 정제 즉, 전처리(pre-processing)
- 결론적으로, 모델의 성능이 중요하다(모델이 어떤 학습방식으로 만들어져있는지에 따라서 다양한 평가방식이 존재)
- 학습방식의 종류(Supervised Learning - 지도학습 , Un-Supervised Learning - 비지도학습 , Reinforcement - 강화 학습)
- 학습방식의 차이점?
- 지도학습은 머신에게 데이터, 정답을 보여주고 학습시키는 방식(예측 - regression , 분류 - classification)
- 비지도학습은 머신에게 데이터만 보여주고 학습시키는 방식(군집화 - clustering)
- 강화학습은 데이터가 없고 스스로 주변에서 데이터를 수집하여 학습하는 방식

모델을 만드는 절차
- 데이터준비(100,000건 정도의 데이터가 필요하고 그래야 모델의 성능을 보장 할 수 있음.)

- 학습데이터를 기반으로 모델(ML 알고리즘이 적용된)을 이용해서 학습을 진행한다. - fit()
- 분류 또는 회귀를  테스트데이터를 이용해서 예측을 수행 - predict()

```python
print('iris 분류모델(ML 알고리즘이적용된)을 이용한 분류예측(classification)을 진행 - ') 
```

```python
print('데이터 분리 - ')
X_train, X_test , y_train , y_test = train_test_split(   iris.data , 
                                                         iris.target , 
                                                         test_size = 0.2 , 
                                                         shuffle   = True , 
                                                         random_state = 200)
```

```python
X_train.shape , X_test.shape  , y_train.shape  , y_test.shape  
```

```python
y_test
```

```python
print('학습 - fit() ')
print('학습을 위한 분류 모델 선정 - ')
print('학습은 train 데이터의 feature, target 이용한다. - ')
tree_model = DecisionTreeClassifier();
tree_model.fit(X_train, y_train)
```

```python
print('분류예측 수행 - predict() ')
print('분류예측 수행에는 테스트데이터의 feature 이용 - ')
y_predict = tree_model.predict(X_test)
print('model answer - ' , y_predict)
print('test  answer - ' , y_test)
```

```python
print('분류된 모델의 예측성능을 확인( 정확도 ) - ')
print('accuracy - ' , accuracy_score(y_test , y_predict))
```

```python
bunseok_feature_frm = titanic_subset_frm[['survived','pclass','sex','age','sibsp','parch','fare','embarked']]
```

```python
bunseok_feature_frm.info()
```

```python
sex_encoding_frm = pd.get_dummies(bunseok_feature_frm['sex'])
sex_encoding_frm 
```

```python
bunseok_feature_frm = pd.concat([sex_encoding_frm, bunseok_feature_frm] , axis = 1)
bunseok_feature_frm.head()
```

```python
embarked_encoding_frm = pd.get_dummies(bunseok_feature_frm['embarked'])
embarked_encoding_frm 
```

```python
bunseok_feature_frm = pd.concat([embarked_encoding_frm, bunseok_feature_frm] , axis = 1)
bunseok_feature_frm.head()
```

```python
bunseok_feature_frm.drop(['sex' , 'embarked'] , axis=1 , inplace=True)
bunseok_feature_frm.head()
```

```python
# bunseok_feature_frm.info()
bunseok_feature_frm.columns
```

```python
features = bunseok_feature_frm[['C', 'Q', 'S', 'female', 'male', 'pclass', 'age', 'sibsp',
       'parch', 'fare']]
target   = bunseok_feature_frm['survived']
```

```python
target
```

```python
X_train, X_test , y_train , y_test = train_test_split(   features , 
                                                         target , 
                                                         test_size = 0.2 , 
                                                         shuffle   = True , 
                                                         random_state = 200)

print('shape - ' , X_train.shape , X_test.shape  , y_train.shape  , y_test.shape  )
```

```python
print('모델 생성')
tree_model = DecisionTreeClassifier();
tree_model.fit(X_train, y_train)

knn_model  = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
```

```python
tree_y_predict = tree_model.predict(X_test)
print('tree model predict - ' , tree_y_predict[:10])
print('tree test  answer  - ' , y_test[:10])
knn_y_predict = knn_model.predict(X_test)
print('knn  model predict - ' , knn_y_predict[:10])
print('knn  test  answer  - ' , y_test[:10])
```

```python
print('분류된 모델의 예측성능을 확인( 정확도 ) - ')
print('tree accuracy - ' , accuracy_score(y_test , tree_y_predict))
print('knn  accuracy - ' , accuracy_score(y_test , knn_y_predict))
```

```python
print('classification report - 평가지표 계산')
print('분류모델에서의 성능평가는 정확도, 정밀도, 재현율, 조화평균 - ') 
print('tree_report = ' )
print()
print(classification_report(y_test , tree_y_predict))
print()
print('knn_report  = ' )      
print()
print(classification_report(y_test , knn_y_predict))
```
