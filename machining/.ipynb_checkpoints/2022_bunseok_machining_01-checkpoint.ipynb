```python
bunseok_feature = titanic_frm[titanic_frm.columns[:8]]
bunseok_feature
```

```python
sex_encoding_frm = pd.get_dummies(bunseok_feature['sex'])
sex_encoding_frm
```

```python
bunseok_feature_frm = pd.concat([sex_encoding_frm , bunseok_feature] , axis = 1)
bunseok_feature_frm
```

```python
embarked_encoding_frm = pd.get_dummies(bunseok_feature['embarked'])
embarked_encoding_frm
```

```python
bunseok_feature_frm = pd.concat([embarked_encoding_frm , bunseok_feature] , axis = 1)
bunseok_feature_frm
```

```python
bunseok_feature_frm.drop(['sex' , 'embarked'] , axis = 1 , inplace = True)
```

```python
bunseok_feature_frm
```

```python
bunseok_feature_frm.info()
```

```python
bunseok_feature_frm
```

```python
features = bunseok_feature_frm[bunseok_feature_frm.columns]
target   = bunseok_feature_frm['survived']
```

```python
features
```

```python
target
```

```python
# 모델링이라고 칸다. 
# 오전 빠져서 뭔지 모르것다 

X_train , X_test , y_train , y_test = train_test_split ( features 
                                                       , target
                                                       , test_size = 0.2
                                                       , shuffle   = True
                                                       , random_state = 200)
print('shqpe - ' , X_train , X_test , y_train , y_test)
```

```python
print('모델 생성')
tree_model = DecisionTreeClassifier() ;
tree_model.fit(X_train , y_train)

knn_model = KNeighborsClassifier() 
knn_model.fit(X_train , y_train)
```

```python
tree_y_predict = tree_model.predict(X_test)
print('tree model answer - ' , tree_y_predict)
print('tree test  answer - ' , y_test.values)

tree_y_predict = knn_model.predict(X_test)
print('knn model answer - ' , tree_y_predict)
print('knn test  answer - ' , y_test.values)
```

```python
# 평가를 한단다
# 뭔말인지 1도 모르것다
# 짜증
# 나

print('분류된 모델의 예측성능을 확인(정확도) - ')
print('tree accuracy - ' , accuracy_score(y_test , tree_y_predict))
```
