```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep, time
```

- By.ID	태그의 id값으로 추출
- By.NAME	태그의 name값으로 추출
- By.XPATH	태그의 경로로 추출
- By.LINK_TEXT	링크 텍스트값으로 추출
- By.PARTIAL_LINK_TEXT	링크 텍스트의 자식 텍스트 값을 추출
- By.TAG_NAME	태그 이름으로 추출
- By.CLASS_NAME	태그의 클래스명으로 추출
- By.CSS_SELECTOR	css선택자로 추출

# 네이버 쇼핑 리뷰

```python
naver_driver = webdriver.Chrome('../diver/chromedriver.exe')
url = 'https://smartstore.naver.com/homgol/products/5811396719?NaPm=ct%3Dlbz9opc8%7Cci%3D4d9ffde4a2433962019bfa58e812dd675582b55c%7Ctr%3Dslcc%7Csn%3D1115439%7Chk%3D015623f62c6cffc394901143d6b8d166c86fe095'
naver_driver.get(url)
```

```python
# 초기페이지에서 상품 리뷰란으로 이동

naver_driver.find_elements(By.CSS_SELECTOR, f'._1k5R-niA93')[1].click()
```

```python
# 페이지당 리뷰 개수 가져오기

len(naver_driver.find_elements(By.CSS_SELECTOR , '._1mUQXqW3YL'))
```

```python
# 리뷰 다음페이지 버튼 XPATH 가 리뷰마다 다른데, 그 이유는 아래 코드의 길이 때문이다.
# 아래 코드는 리뷰 페이지의 수이며, 해당 길이를 '다음'버튼 XPATH 값에 넣어주면 된다.
# 단 임의적으로 코드 실행전 페이지를 건들시 문제가 발생한다 

length = len(naver_driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a'))
length
```

```python
naver_driver.find_elements(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a')[1].text
```

```python
# 200페이지 까지의 리뷰를 가져오기

naver_review_sample_data = []

for n in range(200) :
    for i in range(1 , len(naver_driver.find_elements(By.CSS_SELECTOR , '._1mUQXqW3YL'))+1) :
        data = naver_driver.find_element(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span').text
        if data == '재구매' :
            naver_review_sample_data.append(naver_driver.find_element(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span[2]').text)
        else :
            naver_review_sample_data.append(naver_driver.find_element(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/ul/li[{i}]/div/div/div/div[1]/div/div[1]/div[2]/div/span').text)
        # print('-'*60)
    naver_driver.find_element(By.XPATH, f'//*[@id="REVIEW"]/div/div[3]/div[2]/div/div/a[{length}]').click()
    # 딜레이 시간을 좀 많이 줘야한다.
    sleep(1.5)
    print(f'{n+1}page')
    #print()
```

```python
# 4천개의 리뷰 데이터
# 4만6천 개중 4천개 가져오는데도 3~4분 가까이 걸린 듯 하다.
# 즉 4만개 다가져오면 30분이 걸린다 맙소사 

naver_review_sample_data
```

# 마켓커리 크롤링

```python
market_driver = webdriver.Chrome('../diver/chromedriver.exe')
url = 'https://www.kurly.com/goods/5062329'
market_driver.get(url)    
```

```python
# 상품 후기란으로 이동

market_driver.find_element(By.XPATH , '//*[@id="top"]/div[3]/div[2]/nav/ul/li[3]/a').click()
```

```python
#//*[@id="review"]/section/div[2]/div[6]/article/div/p
market_driver.find_element(By.XPATH , f'//*[@id="review"]/section/div[2]/div[5]/article/div/p').text
```

```python
market_driver.find_elements(By.XPATH , '//*[@id="review"]/section/div[2]/div')[4].text
```

```python
# 페이지당 리뷰 개수를 구하기 위해 가져온 데이터
# class css-169773r e36z05c0 로 가져오면 편하겠지만 이유는 몰라도 안된다.
# 해당 코드는 XPATH로 가져온 리뷰 테이블이며 해당 테이블의 길이 -5 를 해주면 리뷰의 개수를 알 수 있다.

len(market_driver.find_elements(By.XPATH , '//*[@id="review"]/section/div[2]/div')) - 5
```

```python
# 100 페이지 까지 데이터를 가져오자 - faile
# 마켓커리는 일정 데이터 이상 가져오면 막혀버린다.
market_review_sample_data = []
market_review_sample_data_2 = []

for n in range(100) :
    # 리뷰가 5부터 시작하므로 -5값을 지워주자.
    # 리뷰 개수만큼 반복한다.
    for i in range(5 , len(market_driver.find_elements(By.XPATH , '//*[@id="review"]/section/div[2]/div'))) :
        market_review_sample_data_2.append(market_driver.find_element(By.XPATH , f'//*[@id="review"]/section/div[2]/div[{i}]/article/div/p').text)
    print(f'{n} page')
    print()
    sleep(1.5)
    # 리뷰 페이지 이동
    market_driver.find_element(By.XPATH , f'//*[@id="review"]/section/div[2]/div[15]/button[2]').click()
```

```python
# 리뷰 1~2 페이지
market_review_sample_data
```

```python
# 리뷰 3~4 페이지
market_review_sample_data_2
```
