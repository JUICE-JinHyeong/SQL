```python
def file_sample_function(fileName : str , mode : str) :
    print('file path - ' , fileName)
    print('mode      - ' , mode)
    
    try :
        if mode == 'r' :
            file = open(fileName , mode = mode , encoding = 'UTF-8')
            print('type - ' , type(file))
            print('read - \n' , file.read())

        elif mode == 'w' :
            file = open(fileName , mode = mode , encoding = 'UTF-8')
            file.write(input('입력할 내용을 적으세요.\n'))
            
        elif mode == 'a' :
            file = open(fileName , mode = mode , encoding = 'UTF-8')
            file.write('\n- 문서의 마지막 줄입니다. -')
    except FileNotFoundError as f :
        print(str(f))
    
    finally :
        if file != None :
            file.close()
        
```

```python
print('읽기 모드 - ')
file_sample_function('C:/Users/a/data/greeting.txt','r')
print('쓰기 모드 - ')
file_sample_function('C:/Users/a/data/testing.txt' ,'w')
print('추가 모드 - ')
file_sample_function('C:/Users/a/data/testing.txt' ,'a')
```

```python
def file_with_open_function(fileName , mode , encoding) :
    print('file     - ' , fileName)
    print('mode     - ' , mode)
    print('encoding - ' , encoding)
    
    with open(fileName , mode , encoding = encoding) as file :
        data_w = None
        while True :
            data_w = input('내용을 입력하세요. 나가려면 exit 입력하세요.\n')
            
            if  data_w == 'exit' :
                break 
            else :
                file.write(data_w + '\n')
                
            
            
file_with_open_function('C:/Users/a/data/multiline.txt' ,'w' , 'UTF-8')        
```

```python
'''
cnt_words.txt 줄 단위로 읽어서 단어의 길이가 10 이하인 단어들만 카운팅 해본다면?
'''
def cnt_function(fileName , mode , encoding) :    
    cnt = 0
    with open(fileName ,mode , encoding = encoding) as file :
        for line in file :
            word = line.strip('\n')
            print(word)
            if( len(word) <= 10) :
                cnt += 1
                
    print(cnt)
    
```

```python
cnt_function('C:/Users/a/data/cnt_words.txt' , 'r' , 'UTF-8')
```

```python
def isPalindrome() :
    word = input('회문 검사를 위한 단어 압력 : \n')
    middle_idx = len(word)//2
    print('middle_idx - ' , middle_idx)
    
    IsFlag = True
    for idx in range(middle_idx) :
        if word[idx] != word[-1-idx] :
            isFlag = False
            break
    if IsFlag :
        return word
```

```python
'''
네트워크상에서 표준으로 사용되는 파일 형석 - JSON
{ key : value , key : value}
encoding : python(dict) -> JSON
decoding : json         -> python(dict, list) : loads()
'''

import json as j
tmp_dict = {'id' : 'jslim' , 'pwd' : 'jslim'}
str_dict = j.dumps(tmp_dict)
print('type - ' , type(tmp_dict))
print('dict -> json' , j.dumps(tmp_dict))
print('dict -> json' , type(j.dumps(tmp_dict)))
print()
py_dict = j.loads(str_dict)
print('json -> dict' , py_dict)
print('json -> dict' , type(py_dict) , py_dict['id'])
```
