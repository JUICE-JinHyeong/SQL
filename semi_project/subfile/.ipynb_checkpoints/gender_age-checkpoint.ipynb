```python
import pandas as pd
import numpy as np
```

```python
# 연도별 성별 그래프 불러오기
# if문에서 활용하기 위해 '합계' 비교시 True와 False를 반환하는 series 데이터 df_bool 생성
df_a = pd.read_excel('C:/Users/PiGiraffe0/Documents/SQL/semi_project/data/gender_age.xlsx')
df_bool = df_a[['가구주의 연령별']] == '합계'
```

```python
# series.bool() series 형태나 dataframe 형태를 boolean 형태로 바꾸어주며 이 때 값은 한 개여야한다. 
# for 문 안에 넣어서 True 인 값만 추출하여 해당 series의 index 번호인 i를 리스트 lst_data안에 저장한다.

lst_data = []

for i in range(52) :
    is_bool = df_bool.loc[i]
    is_bool = is_bool.bool()
    if is_bool :
        # print(df_a.loc[i])
        lst_data.append(i)
    
lst_data
```

```python
# data 확인 후 , 데이터를 가꾸기 위해 gender_data series 변수를 만든다
gender_data = df_a.loc[lst_data]
gender_data
```

```python
# 열 삭제는 행 삭제와 같은 drop()을 사용하며 이 때 drop 내부에 axis = 1을 적어둬야한다.
gender_data = gender_data.drop(['행정구역별(시군구)'] , axis = 1)

# 지금 오류는 이미 drop으로 해당 칼럼을 삭제해서 칼럼이 없다고 오류가 뜨는 것
```

```python
# 데이터 칼럼명 변경
gender_data = gender_data.rename(index   = {1 : 0 , 18 : 1 , 35 : 2} ,
                   columns = {'가구주의 성별' : 'gender' , '가구주의 연령별' : 'age' , 
                              2015 : '2015' , 2016 : '2016' , 2017 : '2017' , 2018 : '2018' , 2019 : '2019' ,
                              2020 : '2020' , 2021 : '2021'})
```

```python
def to_extract(series , year) :
    index_in = input('인덱스에 들어갈 문자를 입력하세요 : \n' )
    dict_value = { 
        '19세 이하' : series.loc[1:2].sum(axis=0)[year]  ,
        '20 ~ 34세' : series.loc[3:5].sum(axis=0)[year]  ,
        '35 ~ 49세' : series.loc[6:8].sum(axis=0)[year]  ,
        '50 ~ 64세' : series.loc[9:11].sum(axis=0)[year] ,
        '65세 이상' : series.loc[11:].sum(axis=0)[year] 
         }
    df_value = DataFrame(dict_value , index = [index_in])
    return df_value
    
```
