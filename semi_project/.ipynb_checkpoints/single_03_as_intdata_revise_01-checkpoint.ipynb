```python
def to_sharing_value(series) :
    # dataframe - series 데이터를 받아옵니다.
    try :
        mean_data = series.mean()
        # 데이터가 1만으로 나눌 수 있는지 칼럼별 데이터 평균을 구합니다.
        bool_data = None
        # series형은 if문에서 참, 거짓을 표현할 수 없기 때문에 series를 형변환 해주어야합니다.
        # for 내부에서 사용할 변수 입니다. 현재는 초기화를 위한 선언만 존재합니다.
        for value in series :
            # 칼럼별로 루프를 돌려줍니다.
            bool_data = mean_data[[value]] > 10000
            # bool_data에 특정 칼럼의 데이터 평균이 1만보다 큰 지 확인하여 True 나 False를 인자로 갖습니다.
            # 이 때 이 변수는 아직 series 형태입니다.
            bool_data = bool_data.bool()
            # series 형태의 bool_data를 boolean 형태로 형변환 시켜줍니다.
            if bool_data :
                print('{}를 만 단위로 가공합니다.(data/10000)'.format(value))
                series[[value]] = series[[value]] / 10000
                # bool_data가 참인지 거짓인지를 판단하여, 참일 경우 특정 칼럼 데이터 전체를 10000으로 나눕니다.
            else :
                print('{}가 만 단위로 가공되어있습니다.'.format(value))
    except Exception :
        print('error')
    finally :
        return round(series , 1)
        # 소수점 한자리까지 반환합니다.
        # 이 때 전체를 정수형으로 반환하려면 series.astype(int)를 입력하면 됩니다.
```

```python
to_sharing_value(single_03_v1)
```

#### 아래는 참고자료 입니다.

####  dataframe은 반환을 series로 보여주기 때문에 가공에 있어 문제가 생길 수 있다.
-  .to_list() series를 list로 변환
-  series.bool() series를 bool로 변환
-  if 문 비교시 series를 bool로 비교후 선언할 것

#### .astype() 
- dataframe의 형을 특정 형으로 변형한다. 이 때 전체 데이터 형변환을 할 시에 문자와 숫자가 섞여있으면 안되니 주의
- 만일 문자와 숫자가 섞여있다면 dataframe[['칼럼이름, 칼럼이름 ...']] 으로 추출하도록 하자.
- .loc은 이름으로, .iloc은 숫자로 인덱싱 -> 특정 데이터를 찾을때 좋다.
- .loc[index] .iloc[index]
