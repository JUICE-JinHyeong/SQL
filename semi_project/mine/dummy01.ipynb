# 전국 1인가구 현황 지도 시각화

```python
import requests
import json
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
print(folium.__version__)
```

```python
state_geo = 'TL_SCCO_SIG_WGS84.json'
```

```python
single_district = pd.read_excel('./data/single_02.xlsx')
single_district
```

```python
# initialize the map:

m = folium.Map(location=[36,127], tiles="OpenStreetMap", zoom_start=7)
m

m.choropleth(
    geo_data =state_geo,
    name = '행정지역별 1인가구',
    data = single_district,
    columns = ['서울특별시','부산광역시']
    key_on='feature.properties.SIG_CD',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.3,
    color = 'gray',
    legend_name = 'Population')


folium.LayerControl().add_to(m)
```
