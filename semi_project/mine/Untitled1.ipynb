```python
def to_sharing_value(series) :
    # dataframe - series 데이터를 받아옵니다.
    try :
        mean_data = series.mean()
        # 데이터가 1만으로 나눌 수 있는지 칼럼별 데이터 평균을 구합니다.
        bool_data = None
        # 이 변수는 series형은 if문에서 참, 거짓을 표현할 수 없기 때문에 series를 형변환 해주어야합니다.
        # for 내부에서 사용할 bool변수 입니다.
        for value in series :
            # 칼럼별로 루프를 돌려줍니다.
            bool_data = mean_data[[value]] > 10000
            # bool_data에 특정 칼럼의 데이터 평균이 1만보다 큰 지 확인하여 True 나 False를 인자로 갖습니다.
            # 이 때 이 변수는 아직 series 형태입니다.
            bool_data = bool_data.bool()
            # series 형태의 bool_data를 boolean 형태로 형변환 시켜줍니다.
            if bool_data :
                print('{}를 만 단위로 가공합니다.(data/10000)'.format(value))
                series[[value]] = series[[value]] / 10000
                # bool_data가 참인지 거짓인지를 판단하여, 참일 경우 특정 칼럼 데이터 전체를 10000으로 나눕니다.
            else :
                print('{}가 만 단위로 가공되어있습니다.'.format(value))
    except Exception :
        print('error')
    finally :
        return round(series , 1)
        # 소수점 한자리까지 반환합니다.
        # 이 때 전체를 정수형으로 반환하려면 series.astype(int)를 입력하면 됩니다.
```
