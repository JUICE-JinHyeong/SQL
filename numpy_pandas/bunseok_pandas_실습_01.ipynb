```python
# 5 cctv와 인구현황 데이터 파악하기
# 소계를 기준으로 오름차순 정렬
# cctv가 가장 많은 구 확인
# 2014-2016년 cctv 수를 더하고 2013년 이전 cctv 수로 나누어서 최근 3년간 ccrv 증가율을 계산
```

```python
# 소계를 기준으로 오름차순 정렬
cctv_sort = seoul_cctv_frm.sort_values('소계')
seoul_cctv_frm.sort_values('소계').head()
```

```python
# cctv가 가장 많은 구 확인
cctv_sort.iloc[[len(cctv_sort)-1]]
```

```python
# 2014-2016년 cctv 수를 더하고 2013년 이전 cctv 수로 나누어서 최근 3년간 ccrv 증가율을 계산
cctv_upper_frm = round((seoul_cctv_frm['2014년'] + seoul_cctv_frm['2015년'] + seoul_cctv_frm['2016년']) / seoul_cctv_frm['2013년도 이전'] * 100 , 2)
cctv_upper_frm

# 칼럼 추가

seoul_cctv_frm['증가율'] = cctv_upper_frm
seoul_cctv_frm.head()
```

```python
seoul_cctv_frm.sort_values('증가율' , ascending = False).head(1)
```

```python
# 6 서울시 인구 데이터 확인
# drop을 이용해서 합계 행 삭제
# 구별 결측값 확인 - 만약 결측값이 있다면 행 삭제
# 각 구별 전체 인구를 이요하여 구별 외국인 비율과 고령자 비율 계산
# 인구수로 정렬하여 인사이트 찾아보기
seoul_pop_frm.head()
```

```python
# drop을 이용해서 합계 행 삭제
seoul_pop_frm.drop(0 , axis = 0 , inplace = True)
seoul_pop_frm.reset_index(drop=True , inplace = True)
seoul_pop_frm.head()
```

```python
# 구별 결측값 확인 - 만약 결측값이 있다면 행 삭제
seoul_pop_frm.isnull()
seoul_pop_frm.drop(25 , axis = 0 , inplace = True)
seoul_pop_frm
```

```python
#.unique()
# 중복을 제외하고 데이터를 뽑아주는 함수
seoul_pop_frm['구별'].unique()
seoul_pop_frm['구별'].isnull()
seoul_pop_frm[seoul_pop_frm['구별'].isnull()]
```

```python
# 각 구별 전체 인구를 이용하여 구별 외국인 비율과 고령자 비율 계산
foreign_frm = round(seoul_pop_frm['외국인'] / seoul_pop_frm['인구 수'] * 100 , 2)
foreign_frm.head()

older_frm = round(seoul_pop_frm['고령자'] / seoul_pop_frm['인구 수'] * 100 , 2)
older_frm.head()

seoul_pop_frm['외국인 비율'] = foreign_frm
seoul_pop_frm['고령자 비율'] = older_frm

seoul_pop_frm.head()
```

```python
seoul_pop_frm.sort_values('인구 수' , ascending = False).head()
```

```python
# 7 인구대비 cctv 현황을 분석하기 위해서 테이블 병합
```

```python
merge_outer_frm = pd.merge(seoul_cctv_frm , seoul_pop_frm , on = '구별')
merge_outer_frm.head()
```

```python
merge_outer_frm.drop(['2013년도 이전' , '2014년' , '2015년' , '2016년'] , axis = 1 , inplace = True)
merge_outer_frm
```

```python
# 인덱스를 구별로 변환
```

```python
merge_outer_frm.set_index('구별' , inplace = True)
merge_outer_frm
```

#### 상관관계 지수를 확인하는 함수
- -1 ~ 1 사이의 값을 확인
- 0에 가까울수록 상관관계가 약하다.
- -1 or 1dp 가까울수록 상관관계가 강하다.
- np.corrcoef()

```python
print('인구 수 - 소계 \n', np.corrcoef(merge_outer_frm['인구 수'] , merge_outer_frm['소계']))
print()
print('고령자 - 소계 \n', np.corrcoef(merge_outer_frm['고령자'] , merge_outer_frm['소계']))
print()
print('외국인 비율 - 소계 \n', np.corrcoef(merge_outer_frm['외국인 비율'] , merge_outer_frm['소계']))
print()
print('고령자 비율 - 소계 \n', np.corrcoef(merge_outer_frm['고령자 비율'] , merge_outer_frm['소계']))
```
