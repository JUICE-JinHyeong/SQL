```python
ratings_data = pd.read_csv('C:/Users/PiGiraffe0/Documents/SQL/data/pandas/user/ratings.dat' , sep = '::' , header=None)
ratings_data
```

### 02 : 칼럼 생성

```python
# ratings_dat

ratings_dat = ratings_data
ratings_dat.rename(columns = {0 : 'userid' , 1 : 'movieid' , 2 : 'rating' , 3 : 'timestamp'} , inplace =True)
ratings_dat.head()
```

```python
# movies_dat

movies_dat = movies_data
movies_dat.rename(columns = {0 : 'movieid' , 1 : 'title' , 2 :'genres'} , inplace = True)
movies_dat.head()
```

```python
# 3번째 칼럼을 정리하기 위해 아래와 같은 코드 작성

lst = []
for idx in range(len(movies_dat)) :
    if movies_dat['genres'].loc[idx].find('|') != -1 :
        lst.append(movies_dat['genres'].loc[idx][0 : movies_dat['genres'].loc[idx].find('|')])
    else :
        lst.append(movies_dat['genres'].loc[idx])
        
lst

movies_dat['genres'] = lst
movies_dat.head()
```

```python
# users

users_dat = users_data
users_dat.rename(columns = { 0 : 'userid' , 1 : 'gender' , 2 : 'age'} , inplace = True)
users_dat

# 칼럼 3 , 4 제거

users_dat.drop([3 , 4] , axis = 1 , inplace = True)
users_dat.head()
```

### 03 : 공통 칼럼을 이용한 데이터 병합

```python
summaris_dat = pd.merge(pd.merge(movies_dat , ratings_dat , how ='inner') , users_dat)
summaris_dat.info()
summaris_dat
```

### 08 : 여성이 가장 좋아하는 영화 순위

```python
display(summaris_dat)
rating_woman = summaris_dat[summaris_dat['gender'] == 'F'].groupby(['title','gender']).count().sort_values('rating' , ascending = False)[['rating']]
rating_woman.rename(columns = {'rating' : '여성 평가 개수'} , inplace = True)
display(rating_woman)
# 여성만 슬라이싱 , 영화 제목과 성별 그룹화 , 이후 평점 평균 확인
summaris_dat[summaris_dat['gender'] == 'F'].groupby(['title','gender']).agg(np.mean)[['rating']]
# sort 로 순위 확인
rating_point_woman = summaris_dat[summaris_dat['gender'] == 'F'].groupby(['title','gender']).agg(np.mean)[['rating']].sort_values('rating' , ascending = False)
rating_point_woman.rename(columns = {'rating' : '여성 평점 평균'} , inplace = True)
display(rating_point_woman)
```

```python
# 두 df 결합

woman_best_movie_5_frm = pd.concat([rating_woman , rating_point_woman] , axis = 1).head()
woman_best_movie_5_frm
```
