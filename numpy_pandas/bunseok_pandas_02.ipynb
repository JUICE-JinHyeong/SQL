```python
def frminfo(df) :
    print('shape     - ' , df.shape)
    print('size      - ' , df.size)
    print('ndim      - ' , df.ndim )
    print('index     - ' , df.index   , type(df.index))
    print('col index - ' , df.columns , type(df.columns))
    print('values    - ')
    print(df.values  , '\n' , type(df.values))
    print()
    print('data - ')
    display(df)
```

DataFrame : row_index , 데이터 조작 , 인덱스 조작
- loc()  : 라벨값 기반
- iloc() : 정수값 기반

```python
sample_frm = pd.DataFrame(np.arange(10,22).reshape(3,4) , 
                          index   = ['row0'+str(i) for i in range(1,4)] , 
                          columns = ['col0'+str(i) for i in range(1,5)])
sample_frm
frminfo(sample_frm)
```

```python
# 행인덱스의 () 와 [] 차이

print('loc 행 인덱스를 사용하여 행 1개를 선택')
print()
print(sample_frm.loc(0))
print()
print(sample_frm.loc['row01'])
print()
print('-'*60)
print()
print('iloc 행 인덱스를 사용하여 행 1개를 선택')
print(sample_frm.iloc(0))
print()
print(sample_frm.iloc[0])
```

```python
# 행 인덱스 추출

print('행 인덱스를 사용하여 행 여러개 선택')
print('loc 행 인덱스')
print(sample_frm.loc[['row01' , 'row03']])
print()
print('iloc 행 인덱스')
print(sample_frm.iloc[[0,2]])
```

```python
# 행 인덱스 슬라이싱

print('행 인덱스를 사용하여 슬라이싱')
print('loc 행 인덱스')
print(sample_frm.loc['row01' : 'row03'])
print()
print('iloc 행 인덱스')
print(sample_frm.iloc[0:3])

# 주의 - 정수형 슬라이싱시. [0:3] 은 [0] , [1] , [2] 정수형을 가져온다. 
# 마지막 정수형 인덱스는 가져오지 않는다.

print()
print(sample_frm.iloc[0] , type(sample_frm.iloc[0]))

# df 잘라낼시, Series 로 데이터 타입이 바뀐다.
```

```python
# 칼럼 인덱싱 , boolean 리턴
 
print(sample_frm['col01'] > 15)
print()

# 행 인덱싱

print(sample_frm.loc[sample_frm['col01'] > 15])
```

```python
print('행 인덱스와 열 인덱스를 사용하여 행 선택 - loc[행 , 열]')
print('frm[열][행]')
print('첫번째 row 모든 열을 추출')
print()
print('라벨 기반 행 추출')
print(sample_frm.loc['row01' , :])
print()
print('정수 기반 행 추출')
print(sample_frm.iloc[0 , :])

print()
print('-' * 60)
print()

print('특정 컬럼 추출')
print()
print('라벨 기반 추출')
print(sample_frm.loc['row02' , : 'col02'])
print()
print('정수 기반 추출')
print(sample_frm.iloc[0 , [0,2]])
```

```python
print('프레임에서 새로운 열이 아닌 행을 추가한다면 - ')
sample_frm.loc['row04'] = [1 , 2 , 3 , 4] 
print(sample_frm)

# sample_frm.append()
```

```python
print('기존 행을 복사해서 행 추가')
sample_frm.loc['row05'] = sample_frm.loc['row02']
print(sample_frm)
```

```python
print('위와 비슷하게 열 추가')
sample_frm['col05'] = sample_frm['col04']
print(sample_frm)
sample_frm['col06'] = 123
print(sample_frm)
```

```python
# 평균 구하기 및 열 추가
score_frm.loc['평균'] = round(score_frm.mean(),2)
score_frm

# 최진형 학생 영어 점수 수정 및 평균 수정
score_frm['eng']['최진형'] = 90
score_frm.loc['평균'] = round(score_frm.mean(),2)
score_frm

# 조용일 학생의 점수를 데ㅐ이터 프레임으로 추출
score_frm.loc[['조용일']]

# 문승환 학생의 점수를 시리즈 타입으로 추출
score_frm.loc['문승환']
print(type(score_frm.loc['문승환']))

# 오한샘 학생 점수 수정 및 평균 계산
score_frm.loc['오한샘'] = 100
score_frm.loc['평균'] = round(score_frm.mean(),2)
score_frm
```

```python
# 평균 구하기 및 열 추가
score_frm['평균'] = round(score_frm.mean(axis=1),2)
score_frm

# 최진형 학생 영어 점수 수정 및 평균 수정
score_frm.loc['최진형','eng'] = 90
score_frm.loc['평균'] = round(score_frm.mean(),2)
score_frm

# 조용일 학생의 점수를 데ㅐ이터 프레임으로 추출
score_frm.loc[['조용일']]

# 문승환 학생의 점수를 시리즈 타입으로 추출
score_frm.loc['문승환']
print(type(score_frm.loc['문승환']))

# 오한샘 학생 점수 수정 및 평균 계산
score_frm.loc['오한샘'] = 100
score_frm.loc['평균'] = round(score_frm.mean(),2)
score_frm

```
