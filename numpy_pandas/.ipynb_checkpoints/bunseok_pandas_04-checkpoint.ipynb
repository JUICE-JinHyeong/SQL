```python
def frminfo(df) :
    print('shape     - ' , df.shape)
    print('size      - ' , df.size)
    print('ndim      - ' , df.ndim )
    print('index     - ' , df.index   , type(df.index))
    print('col index - ' , df.columns , type(df.columns))
    print('values    - ')
    print(df.values  , '\n' , type(df.values))
    print()
    print('data - ')
    display(df)
```

```python
def seriesinfo(s) :
    print('index - ' , s.index , type(s.index))
    print('value - ' , s.values , type(s.values))
    print()
    print('data - ')
    print(s)
    
```

#### 그룹 분석 및 피봇테이블
- groupby(열 또는 열리스트 | 행 인덱스)
- 그룹연산(size , count , mean, median , min , max , sum ,std , var, quantile , first , last , aggregate , describe , transform

```python
grp_frm = pd.DataFrame({
    '학과'  :  ['DA' , 'AI' , 'AI' , 'DA' , 'AI' ,'AI'],
    '학년'  :  [1,2,3,4,2,3],
    '이름'  :  ['섭섭해' , '최강희' , '김희선' ,'신사임당' , '임꺽정' , '홍길동'],
    '학점'  :  [4.5 , 3.5 , 3.2 , 4.3 , 2.1 , 1.9]
})
frminfo(grp_frm)
```

```python
setosa_grp = species_grp.get_group('setosa')
versicolor = species_grp.get_group('versicolor')
virginica = species_grp.get_group('virginica')
```

```python
def get_ration(feature) :
    return feature.max() / feature.min()
```

```python
species_grp.max()/species_grp.mean()
```

```python
species_grp.agg(get_ration)
```

```python
top_03 = iris_frm.sort_values(by='petal_length' , ascending = False).groupby('species').head(3)
top_03.rename(top_03['species'] , inplace = True)
top_03.drop('species' , axis = 1 , inplace = True)
```

```python
top_03
```

```python
iris_frm.loc[iris_frm.groupby('species').groups['setosa']]
```

```python
# apply 사용하기

def top3_petal_length_func(grp) :
    return grp.sort_values(by='petal_length' , ascending = False)[:3]

# 전치행렬 , .T  , /transpose()

species_grp.apply(top3_petal_length_func).T
```
